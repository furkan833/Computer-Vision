```python
orgken=cv2.imread("a.JPG",0)
orgken = cv2.resize(orgken, (600,600))

#orjinal resim alma 
plt.figure(figsize=(20,20))
plt.subplot(2, 2, 1)
plt.title("Orjinal")
plt.imshow(orgken,cmap="gray")

# threshold belirleme medyana göre 
med_val = np.median(orgken)
print(med_val)

low = int(max(0, (1 - 0.33)*med_val))
high = int(min(255, (1 + 0.33)*med_val))
print(low,high)

# Canny 
orgcan = cv2.Canny(orgken, 60, 120)

plt.subplot(2, 2, 2)
plt.title("Canny")
plt.imshow(orgcan,cmap="gray")

# Median Threshold

orgmedcan = cv2.Canny(orgken, low, high)

plt.subplot(2, 2, 3)
plt.title("Median Canny")
plt.imshow(orgmedcan,cmap="gray")

# blur
blurred_img = cv2.blur(orgken, ksize = (3,3))
blurred_img = cv2.Canny(blurred_img, low, high)

plt.subplot(2, 2, 4)
plt.title("Blur Canny")
plt.imshow(blurred_img,cmap="gray")

```
